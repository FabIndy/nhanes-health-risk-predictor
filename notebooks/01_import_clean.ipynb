```python
# Cellule 1 : import et conf

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 100)
print(" Imports OK - Notebook pr√™t")

```

```python
# Cellule 2 ‚Äî Lecture des 10 fichiers NHANES

demo = pd.read_sas("DEMO_L.xpt", format="xport", encoding="latin1")
diet1 = pd.read_sas("DR1TOT_L.xpt", format="xport", encoding="latin1")
diet2 = pd.read_sas("DR2TOT_L.xpt", format="xport", encoding="latin1")
bmx  = pd.read_sas("BMX_L.xpt", format="xport", encoding="latin1")
chol = pd.read_sas("TCHOL_L.xpt", format="xport", encoding="latin1")
diq  = pd.read_sas("DIQ_L.xpt", format="xport", encoding="latin1")
paq  = pd.read_sas("PAQ_L.xpt", format="xport", encoding="latin1")
alq  = pd.read_sas("ALQ_L.xpt", format="xport", encoding="latin1")
slq  = pd.read_sas("SLQ_L.xpt", format="xport", encoding="latin1")   # üí§ Sommeil
smq  = pd.read_sas("SMQ_L.xpt", format="xport", encoding="latin1")   # üö¨ Tabac

print(" Tous les fichiers ont √©t√© lus avec succ√®s")
print("Taille des bases :")
for name, df in [
    ("demo", demo), ("diet1", diet1), ("diet2", diet2),
    ("bmx", bmx), ("chol", chol), ("diq", diq),
    ("paq", paq), ("alq", alq), ("slq", slq), ("smq", smq)
]:
    print(f"{name:<6} : {df.shape}")
```

```python
# Cellule 3 ‚Äî Fusion finale des 10 datasets NHANES

df = (demo
      .merge(diet1, on="SEQN", how="left")
      .merge(diet2, on="SEQN", how="left")
      .merge(bmx, on="SEQN", how="left")
      .merge(chol, on="SEQN", how="left")
      .merge(diq, on="SEQN", how="left")
      .merge(paq, on="SEQN", how="left")
      .merge(alq, on="SEQN", how="left")
      .merge(slq, on="SEQN", how="left")  
      .merge(smq, on="SEQN", how="left"))  
print(" Fusion finale r√©alis√©e ‚Äî dimensions :", df.shape)
df.head()
```

```python
# Cellule 4 ‚Äî S√©lection finale des 37 variables pertinentes

# Liste des variables retenues
selected_vars = [
    "SEQN", "RIDAGEYR", "RIDRETH3", "RIAGENDR",
    "LBXTC",
    "DIQ010",
    "BMXHT", "BMXBMI", "BMXWT", "BMXWAIST", "BMXARMC",
    "DR1TKCAL", "DR2TKCAL", "DR1TPROT", "DR2TPROT", "DR1TSUGR", "DR2TSUGR",
    "DR1TFIBE", "DR2TFIBE", "DR1TTFAT", "DR2TTFAT", "DR1TALCO", "DR2TALCO",
    "DR1TCARB", "DR2TCARB", "ALQ121",
    "DR1_320Z", "DR2_320Z",
    "PAD680", "PAD790Q", "PAD790U", "PAD800", "PAD810Q", "PAD810U", "PAD820",
    "SLD012", "SMQ020"
]

# On garde uniquement ces colonnes, si elles existent dans le DataFrame
cols_to_keep = [c for c in selected_vars if c in df.columns]
missing_cols = [c for c in selected_vars if c not in df.columns]

df_selected = df[cols_to_keep].copy()

print(f" Variables conserv√©es : {len(cols_to_keep)} / {len(selected_vars)}")
if missing_cols:
    print(" Colonnes manquantes :", missing_cols)
else:
    print("Aucune colonne manquante.")

print("Dimensions du DataFrame filtr√© :", df_selected.shape)
df_selected.head()
```

```python
# Cellule 5 ‚Äî Export du DataFrame s√©lectionn√©
df_selected.to_csv("nhanes_selected_2021_2023.csv", index=False)
print(" Fichier export√© : nhanes_selected_2021_2023.csv")
```
